```python
# Necessary pip installs for Kaggle environment
!pip install transformers optuna openpyxl

import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Dropout, BatchNormalization, concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    precision_score,
    recall_score,
    confusion_matrix,
    classification_report
)
from sklearn.preprocessing import MinMaxScaler
from transformers import BertTokenizer, TFBertModel
import optuna
import matplotlib.pyplot as plt
import seaborn as sns

# Suppress warnings about PyTorch weights
import logging
logging.getLogger("transformers.modeling_tf_utils").setLevel(logging.ERROR)

# Load and preprocess data
file_path = '/kaggle/input/training-data-1/Model Training Pacers (1).xlsx'
data = pd.read_excel(file_path, sheet_name='Sheet1', header=None)
data.columns = ['Material_Name', 'Lead_Time']

# Handle missing values
data.dropna(inplace=True)

# Generate initial class labels based on the first word of Material_Name
data['Class_Name'] = data['Material_Name'].str.split().str[0]  # Extract first word

# Set minimum number of samples per class
min_samples_per_class = 2

# Get class counts
class_counts = data['Class_Name'].value_counts()

# Identify classes to keep
classes_to_keep = class_counts[class_counts >= min_samples_per_class].index

# Filter data to keep only classes with sufficient samples
data = data[data['Class_Name'].isin(classes_to_keep)].reset_index(drop=True)

# Re-factorize the labels after filtering
data['Generalized_Class_Label'], class_names = pd.factorize(data['Class_Name'])
num_classes = len(class_names)
print(f'Number of classes after filtering: {num_classes}')

# Ensure test_size is at least equal to the number of classes
min_test_size = num_classes
total_samples = len(data)
test_size_fraction = min_test_size / total_samples

# Adjust test_size_fraction if it's more than 0.5
test_size_fraction = min(test_size_fraction, 0.5)

# Proceed with train-test split
train_data, val_data = train_test_split(
    data,
    test_size=test_size_fraction,
    random_state=42,
    stratify=data['Generalized_Class_Label']
)

print(f"Adjusted test size fraction: {test_size_fraction}")
print(f"Training samples: {len(train_data)}, Validation samples: {len(val_data)}")

# Tokenize and process text input using batch tokenization
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Tokenize training data
train_encodings = bert_tokenizer(
    train_data['Material_Name'].tolist(),
    padding='max_length',
    truncation=True,
    max_length=128,
    return_tensors='tf'
)
train_text_ids = train_encodings['input_ids']
train_attention_masks = train_encodings['attention_mask']

# Tokenize validation data
val_encodings = bert_tokenizer(
    val_data['Material_Name'].tolist(),
    padding='max_length',
    truncation=True,
    max_length=128,
    return_tensors='tf'
)
val_text_ids = val_encodings['input_ids']
val_attention_masks = val_encodings['attention_mask']

# Prepare lead times as an additional input feature
scaler = MinMaxScaler()
train_lead_times = scaler.fit_transform(train_data['Lead_Time'].values.reshape(-1, 1))
val_lead_times = scaler.transform(val_data['Lead_Time'].values.reshape(-1, 1))

# Convert labels to numpy arrays
train_labels = train_data['Generalized_Class_Label'].values
val_labels = val_data['Generalized_Class_Label'].values

# Custom Keras Layer to Wrap TFBertModel
class BertLayer(tf.keras.layers.Layer):
    def __init__(self, **kwargs):
        super(BertLayer, self).__init__(**kwargs)
        self.bert = TFBertModel.from_pretrained('bert-base-uncased')

    def call(self, inputs):
        input_ids, attention_mask = inputs
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        return outputs.pooler_output

# Function to build the model
def build_model(num_classes, dense_units=64, dropout_rate=0.3):
    input_ids_in = Input(shape=(128,), name='input_ids', dtype='int32')
    attention_masks_in = Input(shape=(128,), name='attention_mask', dtype='int32')
    lead_time_in = Input(shape=(1,), name='lead_time_input', dtype='float32')

    # Get BERT outputs using the custom BertLayer
    bert_outputs = BertLayer()([input_ids_in, attention_masks_in])

    # Additional layers
    x = Dense(dense_units, activation='relu')(bert_outputs)
    x = Dropout(dropout_rate)(x)
    x = BatchNormalization()(x)

    # Integrate lead time input
    lead_time_dense = Dense(32, activation='relu')(lead_time_in)
    x = concatenate([x, lead_time_dense])

    x = Dense(dense_units, activation='relu')(x)
    x = Dropout(dropout_rate)(x)
    x = BatchNormalization()(x)

    output_class = Dense(num_classes, activation='softmax', name='output_class')(x)

    model = Model(inputs=[input_ids_in, attention_masks_in, lead_time_in], outputs=output_class)
    return model

# Build and compile the model
model = build_model(num_classes)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Callbacks for Early Stopping and Model Checkpointing
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2)

# Train the model
history = model.fit(
    [train_text_ids, train_attention_masks, train_lead_times],
    train_labels,
    validation_data=([val_text_ids, val_attention_masks, val_lead_times], val_labels),
    epochs=10,
    batch_size=16,  # Adjust based on your hardware capabilities
    callbacks=[early_stopping, model_checkpoint, reduce_lr]
)

# Visualize Training and Validation Accuracy
plt.figure(figsize=(12, 6))
plt.plot(history.history['accuracy'], label='Training Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='o')
plt.title('Training and Validation Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# Visualize Training and Validation Loss
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss', marker='o')
plt.plot(history.history['val_loss'], label='Validation Loss', marker='o')
plt.title('Training and Validation Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

# Load the best model weights
model = tf.keras.models.load_model('best_model.keras', custom_objects={'BertLayer': BertLayer})

# Evaluate the model on validation data
val_predictions = model.predict([val_text_ids, val_attention_masks, val_lead_times])
val_predictions_labels = np.argmax(val_predictions, axis=1)

# Calculate evaluation metrics
val_accuracy = accuracy_score(val_labels, val_predictions_labels)
val_f1 = f1_score(val_labels, val_predictions_labels, average='weighted', zero_division=0)
val_precision = precision_score(val_labels, val_predictions_labels, average='weighted', zero_division=0)
val_recall = recall_score(val_labels, val_predictions_labels, average='weighted', zero_division=0)
val_confusion_matrix = confusion_matrix(val_labels, val_predictions_labels)

print(f'Validation Accuracy: {val_accuracy}')
print(f'F1 Score: {val_f1}')
print(f'Precision: {val_precision}')
print(f'Recall: {val_recall}')

# Adjust the labels and target names for classification report and confusion matrix
unique_labels = np.unique(np.concatenate((val_labels, val_predictions_labels)))
label_names = [class_names[i] for i in unique_labels]

# Classification Report
report = classification_report(
    val_labels,
    val_predictions_labels,
    labels=unique_labels,
    target_names=label_names,
    zero_division=0
)
print('Classification Report:')
print(report)

# Confusion Matrix Visualization
plt.figure(figsize=(12, 10))
sns.heatmap(
    confusion_matrix(val_labels, val_predictions_labels, labels=unique_labels),
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=label_names,
    yticklabels=label_names
)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()

# Output Data Processing
output_data = val_data.copy()
output_data['Predicted_Class_Label'] = val_predictions_labels
output_data['Predicted_Class_Name'] = [class_names[i] for i in val_predictions_labels]
output_data['Count_of_Lead_Times'] = output_data.groupby('Predicted_Class_Label')['Lead_Time'].transform('count')
output_data['Average_Lead_Time'] = output_data.groupby('Predicted_Class_Label')['Lead_Time'].transform('mean')

# Compute semantic similarity using TF-IDF and cosine similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def compute_average_similarity(group):
    names = group['Material_Name'].tolist()
    if len(names) < 2:
        return 1.0  # Maximum similarity if only one item
    tfidf = TfidfVectorizer().fit_transform(names)
    sim_matrix = cosine_similarity(tfidf)
    n = sim_matrix.shape[0]
    total_sim = sim_matrix.sum() - n  # Exclude self-similarity
    avg_sim = total_sim / (n * (n - 1))
    return avg_sim

average_similarity = output_data.groupby('Predicted_Class_Label').apply(compute_average_similarity)
output_data['Average_Semantic_Similarity'] = output_data['Predicted_Class_Label'].map(average_similarity)

# Compute confidence interval using model's softmax probabilities
val_confidences = np.max(val_predictions, axis=1)
output_data['Confidence_Interval'] = val_confidences

# Save output to CSV
output_columns = [
    'Material_Name',
    'Predicted_Class_Name',
    'Count_of_Lead_Times',
    'Average_Lead_Time',
    'Average_Semantic_Similarity',
    'Confidence_Interval'
]
output_data[output_columns].to_csv('output_classification_results.csv', index=False)

# Hyperparameter Optimization using Optuna
def objective(trial):
    dense_units = trial.suggest_int('dense_units', 32, 128)
    dropout_rate = trial.suggest_float('dropout_rate', 0.1, 0.5)
    learning_rate = trial.suggest_float('learning_rate', 1e-5, 1e-3, log=True)

    model = build_model(num_classes, dense_units=dense_units, dropout_rate=dropout_rate)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    history = model.fit(
        [train_text_ids, train_attention_masks, train_lead_times],
        train_labels,
        validation_split=0.2,
        epochs=5,
        batch_size=16,
        verbose=0
    )
    val_accuracy = history.history['val_accuracy'][-1]
    return val_accuracy

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=10)

print(f'Best Hyperparameters: {study.best_params}')

# Plot the optimization history
optuna.visualization.plot_optimization_history(study)

# Plot parameter importance
optuna.visualization.plot_param_importances(study)

```
